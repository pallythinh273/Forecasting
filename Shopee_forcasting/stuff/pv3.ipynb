```python
import plotly.express as px
import warnings
pd.set_option('display.max_columns',None)
pd.options.display.float_format='{:.0f}'.format
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('ShopeeData_cleaned.csv')
data.head()
```

```python
data_sorted = data.sort_values(by=['itemid', 'Date'])
# Sử dụng phương thức groupby để nhóm dữ liệu theo 'itemid'
grouped = data_sorted.groupby('itemid')

data_sorted['num_sold'] = grouped['historical_sold'].diff()
data_sorted
```

```python
data_sorted = data.sort_values(by=['itemid', 'Date'])

# Sử dụng phương thức groupby để nhóm dữ liệu theo 'itemid'
grouped = data_sorted.groupby('itemid')

# Tính toán sự khác biệt giữa các giá trị 'historical_sold' của các ngày liên tiếp
# data_sorted['num_sold'] = grouped['historical_sold'].diff()

# # Khởi tạo giá trị 'num_sold' của ngày đầu tiên trong mỗi nhóm là giá trị 'historical_sold' ban đầu
# data_sorted['num_sold'] = data_sorted['num_sold'].fillna(method='bfill')
```

```python
data_sorted = data_sorted.drop(columns=['historical_sold'])
data_sorted.head()
```

```python
data_sorted.sort_index(inplace=True)
data_sorted
```

```python
data_day_30 = data_sorted[data_sorted['Date'] == '2024-03-30']
data_day_31 = data_sorted[data_sorted['Date'] == '2024-03-31']
data_day_30.set_index('Date', inplace=True)
data_day_31.set_index('Date', inplace=True)
```

```python
data_train = data_sorted[data_sorted['Date'] < '2024-03-30']
data_test = data_sorted[data_sorted['Date'] >= '2024-03-30']
```

```python
data_train.set_index('Date', inplace=True)
data_test.set_index('Date', inplace=True)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
x_train = data_train.drop(columns=['num_sold'])
y_train = data_train['num_sold']
x_test = data_test.drop(columns=['num_sold'])
y_test = data_test['num_sold']
```

```python
model = XGBRegressor(n_estimators=1000, learning_rate=0.012005)
model.fit(x_train, y_train)
```

### Train từ ngày 10 đến ngày 29, dự đoán tập dữ liệu gồm ngày 30 và 31

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
compare_train29_pred_30_31 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred, 'itemid': data_test['itemid'], 'difference': round(y_test - y_pred).astype(int)})

compare_train29_pred_30_31.to_csv('compare_train29_pred_30_31.csv', index=False)
compare_train29_pred_30_31.head()
```

### sử dụng tập train đến ngày 29 dự đoán ngày 30 và ngày 31

```python
x_test_30 = data_day_30.drop(columns=['num_sold'])
y_test_30 = data_day_30['num_sold']
x_test_31 = data_day_31.drop(columns=['num_sold'])
y_test_31 = data_day_31['num_sold']
```

```python
y_pred_30 = model.predict(x_test_30)
y_pred_31 = model.predict(x_test_31)

mse_30 = mean_squared_error(y_test_30, y_pred_30)
mse_31 = mean_squared_error(y_test_31, y_pred_31)
display(mse_30, mse_31)
```

### MSE ngày 30 là 136 và MSE ngày 31 là 56

```python
compare_train29_pred30 = pd.DataFrame({'itemid': data_day_30['itemid'], 'Actual': y_test_30, 'Predicted': y_pred_30, 'difference': round(y_test_30 - y_pred_30).astype(int)})
compare_train29_pred31 = pd.DataFrame({'itemid': data_day_31['itemid'], 'Actual': y_test_31, 'Predicted': y_pred_31, 'difference': round(y_test_31 - y_pred_31).astype(int)})

compare_train29_pred30.to_csv('compare_train29_pred30.csv', index=False)
compare_train29_pred31.to_csv('compare_train29_pred31.csv', index=False)
```

### Train từ ngày 10 đến ngày 30, dự đoán ngày 31

```python
data_train = data_sorted[data_sorted['Date'] < '2024-03-31']
data_test = data_sorted[data_sorted['Date'] >= '2024-03-31']

data_train.set_index('Date', inplace=True)
data_test.set_index('Date', inplace=True)

x_train = data_train.drop(columns=['num_sold'])
y_train = data_train['num_sold']
x_test = data_test.drop(columns=['num_sold'])
y_test = data_test['num_sold']

model = XGBRegressor(n_estimators=1000, learning_rate=0.012005)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
mse
```

```python
compare_4 = pd.DataFrame({'itemid': data_test['itemid'], 'Actual': y_test, 'Predicted': y_pred, 'difference': round(y_test - y_pred).astype(int)})
compare_4.to_csv('compare_train30_pre31.csv')
compare_4.head(10)
```
